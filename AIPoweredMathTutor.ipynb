```python
# reading the Maths word problems dataset.
maths_word_problems_df = pd.read_csv("/content/math_output_file_pruned.csv")

# displaying the first 5 rows of our dataset
maths_word_problems_df.head()
```

**Description:**
This code block is crucial for enhancing math problem-solving skills using a generative AI model, aligning with the project's objectives as follows:

- **Initialize Training Data:** The `training_data` list stores example conversations for training the generative AI model, ensuring it learns to interact conversationally with students.
- **Define System's Role:** `system_message` sets the AI's role as a "school level math assistant," focusing on understanding and providing clear, step-by-step solutions, vital for interactive and personalized learning.
- **Prepare Example Conversations:** The `prepare_example_conversation` function simulates interactions between a student and the AI, capturing the question and providing a step-by-step solution.
- **Demonstrate Functionality:** Using `pprint` to print the conversation structure for the first problem in `maths_word_problems_df` shows how the AI interacts, aiding in model training and ensuring solutions align with educational objectives.

This code supports delivering scalable, adaptable, and engaging math education through generative AI, emphasizing fairness and inclusivity in AI responses.


```python
training_data = []

system_message = "You are a school level math assistant. You have to study the problems provided and come up with solutions in a lucid way."


def prepare_example_conversation(row):
    messages = []
    messages.append({"role": "system", "content": system_message})

    messages.append({"role": "user", "content": row["question"]})

    messages.append({"role": "assistant", "content": row["answer"]})

    return {"messages": messages}

pprint(prepare_example_conversation(maths_word_problems_df.iloc[0]))
```

Let's now do this for a subset of the dataset to use as our training data.

## Data Preparation for Model Training

**Selecting Training Data:**
- `training_df = maths_word_problems_df.loc[0:100]` selects the first 100 rows from the math word problems dataset.
- This subset trains the AI model with a variety of problems, easing it into the full dataset and supporting manageable, iterative improvements.

**Transforming Data for Model Training:**
- `training_df.apply(prepare_example_conversation, axis=1).tolist()` applies `prepare_example_conversation` to each row.
- This likely structures math problems and solutions into a format the AI model can process, enhancing its ability to provide accurate solutions.

**Previewing Training Examples:**
- The loop `for example in training_data[:5]:` prints the first five transformed training examples.
- This step ensures data quality and format consistency before full-scale training, aligning with the project's goal of enhancing math education outcomes.


```python
# using the first 100 rows of the dataset for training
training_df = maths_word_problems_df.loc[0:100]

# applying the prepare_example_conversation function to each row of the training_df
training_data = training_df.apply(prepare_example_conversation, axis=1).tolist()

for example in training_data[:5]:
    print(example)
```

In addition to training data, we can also **optionally** provide validation data, which will be used to make sure that the model does not overfit your training set.

```python
validation_df = maths_word_problems_df.loc[101:200]
validation_data = validation_df.apply(prepare_example_conversation, axis=1).tolist()
```

We then need to save our data as `.jsonl` files, with each line being one training example conversation.


- `write_jsonl` serializes a list of dictionaries into JSON Lines format and writes it to a file.
- JSON Lines format enables efficient storage and processing of large datasets.
- Crucial for managing datasets in the project to fine-tune a generative AI model for math word problems, ensuring data consistency, training efficiency, scalability, and interoperability.


```python
def write_jsonl(data_list: list, filename: str) -> None:
    with open(filename, "w") as out:
        for ddict in data_list:
            jout = json.dumps(ddict) + "\n"
            out.write(jout)
```

**Prepare training dataset:**   
- Set `training_file_name` to "tmp_maths_word_problems_finetune_training.jsonl" for storing training data.  
- Write training data to `training_file_name` using `write_jsonl`.
- Prepare validation dataset:
    - Set `validation_file_name` to "tmp_maths_word_problems_finetune_validation.jsonl" for storing validation data.
    - Write validation data to `validation_file_name` using `write_jsonl`.
- Crucial for fine-tuning the generative AI model for school-level math word problems, enabling efficient storage and processing of training and validation datasets.


```python
training_file_name = "tmp_maths_word_problems_finetune_training.jsonl"
write_jsonl(training_data, training_file_name)

validation_file_name = "tmp_maths_word_problems_finetune_validation.jsonl"
write_jsonl(validation_data, validation_file_name)
```

This is what the first 5 lines of our training `.jsonl` file look like:

- **Quality Check**: Allows verification of math word problems before fine-tuning the model.

### Upload files

Upload the files to our `Files` endpoint to be used by the fine-tuned model.


- **Upload Datasets**: Upload training and validation datasets for model fine-tuning.
- **Training Dataset**: Read and upload the training dataset (training_file_name) for diverse math problems.
- **Validation Dataset**: Read and upload the validation dataset (validation_file_name) for testing model performance.
- **File IDs**: Store and print file IDs for tracking and managing within OpenAI ecosystem.


```python
with open(training_file_name, "rb") as training_fd:
    training_response = client.files.create(
        file=training_fd, purpose="fine-tune"
    )

training_file_id = training_response.id

with open(validation_file_name, "rb") as validation_fd:
    validation_response = client.files.create(
        file=validation_fd, purpose="fine-tune"
    )
validation_file_id = validation_response.id

print("Training file ID:", training_file_id)
print("Validation file ID:", validation_file_id)
```

## Fine-tuning

Creating our fine-tuning job with the generated files and an optional suffix to identify the model. The response will contain an `id` which can be uses to retrieve updates on the job.



- **Initiate Fine-Tuning**: Start fine-tuning the generative AI model for solving math word problems.
- **Model Selection**: Use the GPT-3.5-turbo architecture known for efficient handling of NLP tasks.
- **Training and Validation**: Fine-tune the model with specified training and validation files to learn from diverse math problems.
- **Job Creation**: Create a fine-tuning job with a specific suffix for tracking and management.
- **Track Job Progress**: Monitor job progress using the job ID and status for real-time updates.


```python
response = client.fine_tuning.jobs.create(
    training_file=training_file_id,
    validation_file=validation_file_id,
    model="gpt-3.5-turbo",
)

job_id = response.id

print("Job ID:", response.id)
print("Status:", response.status)
```

#### Check job status

Make a `GET` request to the `https://api.openai.com/v1/alpha/fine-tunes` endpoint to list our alpha fine-tune jobs. In this instance , check that the ID we got from the previous step ends up as `status: succeeded`.

Once it is completed, we can use the `result_files` to sample the results from the validation set, and use the ID from the `fine_tuned_model` parameter to invoke our trained model.


- **Monitor Job**: Retrieve details about the fine-tuning job's progress and status.
- **Job Information**: Display job ID, status, and trained tokens for tracking and optimization.


```python
response = client.fine_tuning.jobs.retrieve(job_id)

print("Job ID:", response.id)
print("Status:", response.status)
print("Trained Tokens:", response.trained_tokens)
```

We can track the progress of the fine-tune with the events endpoint. We can rerun the cell below a few times until the fine-tune is ready.


- **Track Progress**: Retrieve and display events to monitor the fine-tuning job's progress.
- **Retrieve Events**: Get a list of events associated with the job using `client.fine_tuning.jobs.list_events(job_id)`.
- **Reverse Order**: Reverse the events to display the most recent updates first for easier monitoring.
- **Print Messages**: Iterate through events and print their messages for real-time updates.


```python
response = client.fine_tuning.jobs.list_events(job_id)

events = response.data
events.reverse()

for event in events:
    print(event.message)
```

Now that it's done, we can get a fine-tuned model ID from the job:


- **Retrieve Fine-Tuned Model ID**: Get the ID of the fine-tuned AI model for math problem-solving.
- **Error Handling**: Raise an error if the model ID is not found, indicating an incomplete fine-tuning process.
- **Integration**: Use the model ID for integrating the AI model into the system.


```python
response = client.fine_tuning.jobs.retrieve(job_id)
fine_tuned_model_id = response.fine_tuned_model

if fine_tuned_model_id is None:
    raise RuntimeError("Fine-tuned model ID not found. Your job has likely not been completed yet.")

print("Fine-tuned model ID:", fine_tuned_model_id)
```

## Inference

The last step is to use our fine-tuned model for inference. Similar to the classic `FineTuning`, we simply call `ChatCompletions` with our new fine-tuned model name filling the `model` parameter.


- **Select Subset**: Select a subset of math word problems (201-300) for testing the model's performance.
- **Choose Test Problem**: Select the sixth problem (index 5) from the subset for interaction with the model.
- **Create Test Messages**: Initialize a list for messages simulating a conversation between the system and a user. Add system and user messages to the list.
- **Display Messages**: Use pprint to display the test messages for clarity and understanding of the interaction flow.


```python
# Question 1

test_df = maths_word_problems_df.loc[201:300]
test_row = test_df.iloc[0]
test_messages = []
test_messages.append({"role": "system", "content": system_message})

test_messages.append({"role": "user", "content": test_row["question"]})

pprint(test_messages)
```

- **Generate Solution**: Use the fine-tuned AI model to solve math word problems.
- **Model Selection**: Specify the fine-tuned model (fine_tuned_model_id) trained on diverse math problems.
- **Input Problems**: Provide math word problems (test_messages) for the model.
- **Output Control**: Set temperature=0 for consistent outputs and max_tokens=500 for concise solutions.
- **Display Solution**: Print the AI-generated solution for immediate feedback.


```python
response = client.chat.completions.create(
    model=fine_tuned_model_id, messages=test_messages, temperature=0, max_tokens=500
)
print(response.choices[0].message.content)
```
